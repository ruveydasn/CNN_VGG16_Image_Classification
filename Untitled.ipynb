```python
import tensorflow
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.applications.vgg16 import VGG16 
from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions
from PIL import Image 
import matplotlib.pyplot as plt
import numpy as np
```

```python
train_files_path = "sports_1/train/"
test_files_path = "sports_1/test/"


img = load_img(test_files_path + "golf/1.jpg")
```

```python
print(img_to_array(img).shape)
```

```python
plt.imshow(img)
plt.show()
```

```python
train_data = ImageDataGenerator().flow_from_directory(train_files_path,target_size = (224,224))
test_data = ImageDataGenerator().flow_from_directory(test_files_path,target_size = (224,224))
```

```python
numberOfSportsTypes = 19
```

```python
vgg = VGG16()
```

```python
vgg_layers = vgg.layers
print(vgg_layers)
```

```python
vggmodel_layersize_tobe_used = len(vgg_layers) - 1

model = Sequential()
for i in range(vggmodel_layersize_tobe_used):
    model.add(vgg_layers[i])
```

```python
for layers in model.layers:
    layers.trainable = False
```

```python
model.add(Dense(numberOfSportsTypes, activation="softmax"))

print(model.summary())
```

```python
model.compile(loss = "categorical_crossentropy",
              optimizer = "rmsprop",
              metrics = ["accuracy"])
```

```python
batch_size = 4 

model.fit_generator(train_data,
                           steps_per_epoch=400//batch_size,
                           epochs= 3, 
                           validation_data=test_data,
                           validation_steps= 200//batch_size)
```
